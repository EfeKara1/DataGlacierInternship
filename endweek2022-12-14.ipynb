```python
def grap_col_names(dataframe, cardinal_threshold=10):
    """
    Summarises a dataframes categorical and numerical variables.
    Parameters
    ----------
        categorical_threshold: Threshold to define a numerical variable as a categorical variable.
        cardinal_threshold: Threshold to define a categorical variable as a numerical variable.
    Returns
    -------
    Returns lists of columns.
    """
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O" ]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cardinal_threshold
                   and dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > cardinal_threshold
                   and dataframe[col].dtypes == "O"]
    final_cat_cols = list(set(cat_cols + num_but_cat) - set(cat_but_car))
    final_num_cols = list(set(num_cols) - set(num_but_cat))

    print("Observations         : ", dataframe.shape[0])
    print("Variables            : ", dataframe.shape[1])
    print("Categorical Variables: ", final_cat_cols)
    print("Numerical Variables  : ", final_num_cols)
    print("Cat_but_car Variables: ", cat_but_car)
    print("Num_but_cat Variables: ", num_but_cat)

    return final_cat_cols, final_num_cols, num_but_cat, cat_but_car
```

## RFM Metrics for Yellow Cab Firm

```python
df_yellow["Date of Travel"].max()

today_date = dt.datetime(2018, 12, 31)

rfm_yellow = df_yellow.groupby('Customer ID').agg({'Date of Travel': lambda date: (today_date - date.max()).days,
                                     'Transaction ID': lambda num: len(num),
                                     'Price Charged': lambda PriceCharged: PriceCharged.sum()})
```

```python
rfm_yellow.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
rfm_yellow = rfm_yellow[(rfm_yellow["Monetary"]) > 0 & (rfm_yellow["Frequency"] > 0)]
```

## Calculating RFM Scores for yellow cab firm

```python
rfm_yellow["RecencyScore"] = pd.qcut(rfm_yellow['Recency'], 5, labels=[5, 4, 3, 2, 1])
```

```python
rfm_yellow["FrequencyScore"] = pd.qcut(rfm_yellow['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm_yellow["MonetaryScore"] = pd.qcut(rfm_yellow['Monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm_yellow["RFM_SCORE"] = (rfm_yellow['RecencyScore'].astype(str) +
                    rfm_yellow['FrequencyScore'].astype(str) +
                    rfm_yellow['MonetaryScore'].astype(str))
```

## Naming & Analysing RFM Segments for yellow cab firm dataset

```python
rfm_yellow['Segment'] = rfm_yellow['RecencyScore'].astype(str) + rfm_yellow['FrequencyScore'].astype(str)
```

```python
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At_Risk',
    r'[1-2]5': 'Cant_Loose',
    r'3[1-2]': 'About_to_Sleep',
    r'33': 'Need_Attention',
    r'[3-4][4-5]': 'Loyal_Customers',
    r'41': 'Promising',
    r'51': 'New_Customers',
    r'[4-5][2-3]': 'Potential_Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm_yellow['Segment'] = rfm_yellow['Segment'].replace(seg_map, regex=True)
```

```python
rfm_yellow[["Segment", "Recency", "Frequency", "Monetary"]].groupby("Segment").agg(["mean", "count"])
```
